```sql
CREATE TABLE IF NOT EXISTS  quicksilver.bronze.tab_customers 
AS 
SELECT *, _metadata.file_path AS filePath, _metadata.file_modification_time AS fileModificationTime FROM json.`/Volumes/quicksilver/rawdata/webdata/customers/`
```

```sql
SELECT COUNT(*)
,COUNT(customer_id) 
,COUNT(email)
,COUNT(telephone)
,COUNT(member_since)
FROM quicksilver.bronze.tab_customers
```
